```python
from selenium.webdriver import Chrome
from selenium.webdriver.common.by import By
from selenium.webdriver.common.action_chains import ActionChains
import time

# disease string
str = 'type 2 diabetes'
# Set desired min and max values
desired_min = 0.25
desired_max = 0.75
# browser driver
driver = Chrome()

# url of website
url = "https://dev.parse-health.org/shiny/ARCH/"
driver.get(url)
# sidebar button
button = driver.find_element(By.CLASS_NAME, 'sidebar-toggle')
button.click()
time.sleep(30)
# keyword input box
input_box = driver.find_element(By.ID, 'side-searchbox2_text') 
input_box.clear()
input_box.send_keys(str) 
# Search Button
button = driver.find_element(By.ID, 'side-searchbox2_search') 
button.click()

# find range slider handle
range_slider_handle_from = driver.find_element(By.CSS_SELECTOR, '.irs-handle.from')
# Extract the minimum and maximum values of the slider
range_slider = driver.find_element(By.CSS_SELECTOR, '.irs-hidden-input')
slider_min = float(range_slider.get_attribute('data-min'))
slider_max = float(range_slider.get_attribute('data-max'))
def adjust_handle(driver, handle, value_element, target_value, slider_min, slider_max):
    
    def get_current_value():
        return float(driver.execute_script("return parseFloat(arguments[0].innerHTML);", value_element))
    left, right = 0, 50
    current_value = get_current_value()
    while abs(current_value - target_value) >= 0.01:
        mid = (left + right) / 2
        
        if current_value < target_value:
            offset = mid
            left = mid
        else:
            offset = -mid
            right = mid

        actions = ActionChains(driver)
        actions.move_to_element(handle)
        actions.click_and_hold()
        actions.move_by_offset(offset, 0)
        actions.release()
        actions.perform()
        current_value = get_current_value()
range_slider_value_from = driver.find_element(By.CSS_SELECTOR, '.irs-from')
range_slider_value_to = driver.find_element(By.CSS_SELECTOR, '.irs-to')
# Adjust the min handle
adjust_handle(driver, range_slider_handle_from, range_slider_value_from, desired_min, slider_min, slider_max)
# Adjust the max handle
range_slider_handle_to = driver.find_element(By.CSS_SELECTOR, '.irs-handle.to')
adjust_handle(driver, range_slider_handle_to, range_slider_value_to, desired_max, slider_min, slider_max)

time.sleep(5)
rows = driver.find_elements(By.TAG_NAME,'tr')


for row in rows:
    cols = row.find_elements(By.TAG_NAME,'td')
    for col in cols:
        col.click()
        time.sleep(5)
        button = driver.find_element(By.ID,'side-gobutton')
        button.click()
        time.sleep(5)
        tab = driver.find_element(By.CSS_SELECTOR, 'a[data-value="Table of connected nodes"]')
        tab.click()
        time.sleep(5)
        button = driver.find_element(By.ID,'btn-downloadData')
        button.click()
    
driver.quit()
```
